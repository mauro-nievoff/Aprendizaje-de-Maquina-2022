```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x_dataset_1 = [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]
y_dataset_1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]

x_dataset_2 = [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]
y_dataset_2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]

x_dataset_3 = [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]
y_dataset_3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]

x_dataset_4 = [8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 19.0, 8.0, 8.0, 8.0]
y_dataset_4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
dataset_1 = pd.DataFrame({'x': x_dataset_1, 'y': y_dataset_1, 'fuente': 'dataset_1'})
dataset_2 = pd.DataFrame({'x': x_dataset_2, 'y': y_dataset_2, 'fuente': 'dataset_2'})
dataset_3 = pd.DataFrame({'x': x_dataset_3, 'y': y_dataset_3, 'fuente': 'dataset_3'})
dataset_4 = pd.DataFrame({'x': x_dataset_4, 'y': y_dataset_4, 'fuente': 'dataset_4'})
```

```python
dato_a = pd.DataFrame({'x': [16], 'y': [8], 'fuente': 'dato_a'})
dato_b = pd.DataFrame({'x': [8], 'y': [6], 'fuente': 'dato_b'})
dato_c = pd.DataFrame({'x': [16], 'y': [10], 'fuente': 'dato_c'})
dato_d = pd.DataFrame({'x': [16], 'y': [6], 'fuente': 'dato_d'})

nuevos_datos = [dato_a, dato_b, dato_c, dato_d]
```

```python
def agregar_datos(dataset, evos_datos):

  for dato in nuevos_datos:

    dataset = pd.concat([dataset, dato], ignore_index = True)

  return dataset
```

```python
dataset_1 = agregar_datos(dataset_1, nuevos_datos)
dataset_2 = agregar_datos(dataset_2, nuevos_datos)
dataset_3 = agregar_datos(dataset_3, nuevos_datos)
dataset_4 = agregar_datos(dataset_4, nuevos_datos)

datasets = [dataset_1, dataset_2, dataset_3, dataset_4]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,12))

for i, dataset in enumerate(datasets):

  sns.scatterplot(x = 'x', y = 'y', data = dataset, hue = 'fuente', ax=axes[i//2,i%2]).set(yticklabels=[], ylabel=None)

  axes[i//2,i%2].title.set_text('Dataset ' + str(i + 1))
```

A partir de los gr√°ficos se puede determinar qu√© punto corresponde con cada dataset:
- El dato_b corresponde al dataset 4 ya que es el √∫nico dato nuevo con x = 8, que es el valor que toma x en casi todos los datos del dataset 4.
- El dato_d corresponde al dataset 2, ya que la distribuci√≥n de los datos en este dataset dibuja una par√°bola, y el dato_d se encuentra dentro del recorrido de dicha par√°bola.
- El dato_c corresponde al dataset 3, ya que la distribuci√≥n de los datos en este dataset dibuja una l√≠nea recta, y el dato_c es el que se encuentra m√°s cerca del recorrido de dicha recta.
- Finalmente, el dato_a corresponde al dataset 1 por descarte. Analizando s√≥lo el dataset 1, el dato_b y el dato_c tambi√©n parecen buenas opciones, pero estos datos fueron asignados a otros datasets.
